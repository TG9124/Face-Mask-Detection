```python
import torch
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained(r"runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16)
```

```python
pipe = pipe.to("cuda")
```

```python
def image_generator(prompt="A director shooting an important scene of his movie.The mood of the scene is exciting and inspiring."):
  template = f"Generate a comic book page based on the given scene description below.It should like a production ready comic book. The dialogues should be in enlgish. {prompt}"
  image = pipe(template).images[0]
  return image
```

```python
prompt = input("Prompt : ")
image_generator(prompt)
```
